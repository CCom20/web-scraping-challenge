```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

### NASA Mars News

```python
# Get response from website
response = requests.get('https://mars.nasa.gov/news/')
print(response)
```

```python
# Set up the text variable 'res' and print to check
res = response.text
print(res)
```

```python
# Set up BeautifulSoup object and print to check
soup = bs(res, 'html.parser')
print(soup.prettify())
```

#### NOTE
It looks as though this site is only letting scraping for July and earlier stories. I've tried several times to get November stories but it does not work. Each time I request the URL, I only receive news for July and earlier.

```python
# Get the news title and print to make sure it's correct
news_title = soup.select('div.content_title')[0].text.strip()
news_title
```

```python
# Get the news paragraph, print to make sure it's correct
news_para = soup.select('div.rollover_description_inner')[1].text.strip()
news_para
```
